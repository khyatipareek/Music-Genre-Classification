```python
import pandas as pd
tracks=pd.read_csv("https://raw.githubusercontent.com/stujenn/genre-classifier/master/track_data.csv")


```

```python
#Splitting the dataset by genre for exploratory data analysis
hh_features = tracks[tracks['genre'] == 'Hip-Hop']
cl_features = tracks[tracks['genre'] == 'Classical']
```

```python
# summary statistics of data with genre=hip-hop
print('Hip-Hop:')
hh_features.describe().round(2)
```

```python
# summary statistics of data with genre=classical
print('Classical:')
cl_features.describe().round(2)
```

```python
#checking for collinearity among features by creating a correlation matrix
corr_metrics = tracks.drop(['key', 'mode', 'duration_ms', 'time_signature', 'target'], axis=1).corr()
corr_metrics.style.background_gradient()
```

```python
#principal component analysis for feature reduction
#step 1: normalizing the data 
features = tracks.drop(['genre', 'id', 'type', 'id', 'uri', 'track_href', 'analysis_url', 'target'], axis=1)
labels = tracks['genre']

from sklearn.preprocessing import StandardScaler

# Scale the features
scaler = StandardScaler()
scaled_train_features = scaler.fit_transform(features)
```

```python
# PCA is performed with chosen number of components n=8
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Get our explained variance ratios from PCA using all features
pca = PCA()
pca.fit(scaled_train_features)
exp_variance = pca.explained_variance_ratio_

# Calculate the cumulative explained variance
cum_exp_variance = np.cumsum(exp_variance)

n_components = 8 #since at 0.95 threshold, n=8 will be most effective
pca = PCA(n_components)
pca.fit(scaled_train_features)
pca_projection = pca.transform(scaled_train_features)
```

```python
%matplotlib inline
# Plot the cumulative explained variance and draw a dashed line at 0.95.
fig, ax = plt.subplots()
ax.plot(range(13), cum_exp_variance)
ax.axhline(y=0.95, linestyle='--')
```

```python
#splitting PCA projected data into training and testing models
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression #LOGISTIC REGRESSION
from sklearn.tree import DecisionTreeClassifier     #DECISION TREE

# Split our data
train_features, test_features, train_labels, test_labels = train_test_split(pca_projection,labels)
```

```python
#LOGISTIC REGRESSION
#TRAIN MODEL
logreg = LogisticRegression(solver='lbfgs')
logreg.fit(train_features, train_labels)
#TEST MODEL
pred_labels_logit = logreg.predict(test_features)
```

```python
#DECISION TREE
# TRAIN MODEL
tree = DecisionTreeClassifier()
tree.fit(train_features, train_labels)
# TEST MODEL
pred_labels_tree = tree.predict(test_features)
```

```python
#Comparing decision tree and logistic regression classifiers
from sklearn.metrics import classification_report
class_rep_tree = classification_report(test_labels, pred_labels_tree)
class_rep_log = classification_report(test_labels, pred_labels_logit)

print("Logistic Regression: \n", class_rep_log)
print("Decision Tree: \n", class_rep_tree)
```

```python
#Cross-Validation- for better accuracy
from sklearn.model_selection import KFold, cross_val_score
kf = KFold(n_splits=10)

tree = DecisionTreeClassifier()
logreg = LogisticRegression(solver='lbfgs')

tree_score = cross_val_score(tree, pca_projection, labels, cv=kf)
logit_score = cross_val_score(logreg, pca_projection, labels, cv=kf)

print("Decision Tree:", np.mean(tree_score))
print("Logistic Regression:", np.mean(logit_score))
```
